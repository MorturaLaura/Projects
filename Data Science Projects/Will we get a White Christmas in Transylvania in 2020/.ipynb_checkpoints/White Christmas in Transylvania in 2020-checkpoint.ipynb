## Cluj Napoca

```python
df_Cluj_Napoca.head(32)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_Cluj = df_Cluj_Napoca[["AO_Dec", "NINO3.4_DEC"]]
y_Cluj = df_Cluj_Napoca["CLUJ NAPOCA"]
```

```python
X_train_Cluj, X_test_Cluj, y_train_Cluj, y_test_Cluj = train_test_split(X_Cluj, y_Cluj, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_Cluj = LogisticRegression()
```

```python
model_Cluj.fit(X_train_Cluj, y_train_Cluj)
```

```python
model_Cluj.score(X_test_Cluj, y_test_Cluj)
```

```python
X_test_Cluj
```

```python
model_Cluj.predict_proba(X_test_Cluj)
```

```python
model_Cluj.predict_proba([[0, -1.5]])
```

```python
model_Cluj.predict([[0, -1.5]])
```
